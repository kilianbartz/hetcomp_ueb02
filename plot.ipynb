```python
import matplotlib.pyplot as plt

aufgabe1 = [435, 421, 1013, 5.3, 83]
aufgabe3 = [52,51,113.6, .5, 10]
aufgabe4 = [171,188,416, 6, 19]

# create bar chart
fig, ax = plt.subplots()
bar_width = 0.2
bar1 = ax.bar([1,2,3,4,5], aufgabe1, bar_width, label='Sequenziell')
bar3 = ax.bar([1+bar_width,2+bar_width,3+bar_width,4+bar_width,5+bar_width], aufgabe3, bar_width, label='Multicore')
bar4 = ax.bar([1+2*bar_width,2+2*bar_width,3+2*bar_width,4+2*bar_width, 5+2*bar_width], aufgabe4, bar_width, label='GPU')
# no x-axis labels or ticks
ax.set_xticks([])
ax.set_xticklabels([])
ax.set_ylabel('Zeit in s')
ax.set_title('Vergleich der sequenziellen, Multicore und GPU Implementierung')
plt.legend()
plt.show()
```
